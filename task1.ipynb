```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lineplt
x = np.linspace(0, 10, 200)
y = np.sin(x)
plt.figure(figsize=(8,4))
plt.plot(x, y, label='sin(x)', linewidth=2)
plt.title('Line plot â€” sine wave')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#scatterplt
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
sizes = 100 * np.random.rand(100)
colors = np.random.rand(100)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6)
plt.title('Scatter plot â€” random points')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='color scale')
plt.show()
```

```python
#barplt
labels = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 12]
plt.figure(figsize=(6,4))
plt.bar(labels, values)
plt.title('Bar chart â€” categorical values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#hiatogramplt
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram â€” normal distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#pieplt
sizes = [15, 30, 45, 10]
labels = ['Group A', 'Group B', 'Group C', 'Group D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie chart â€” category shares')
plt.axis('equal') # keeps pie circular
plt.show()
```

```python
#boxplt
data = [np.random.normal(loc, 1, 200) for loc in [0, 2, 4]]
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Boxplot â€” compare distributions')
plt.show()
```

```python
#heatmap matrixplt
matrix = np.random.rand(10, 10)
plt.imshow(matrix, aspect='auto')
plt.colorbar()
plt.title('Heatmap â€” random matrix')
plt.show()
```

```python
#subplotplt
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
axes[0,0].plot(x, np.sin(x)); axes[0,0].set_title('sin')
axes[0,1].plot(x, np.cos(x)); axes[0,1].set_title('cos')
axes[1,0].hist(np.random.randn(500), bins=20); axes[1,0].set_title('hist')
axes[1,1].scatter(np.random.rand(50), np.random.rand(50)); axes[1,1].set_title('scatter')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sns.set_theme(style='whitegrid')
```

```python
#scattersns
tips = sns.load_dataset('tips')
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', style='time')
plt.title('Scatter â€” tips dataset')
plt.show()
```

```python
#lineplotsns
fmri = sns.load_dataset('fmri')
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event')
plt.title('Line plot â€” fmri example')
plt.show()
```

```python
#barplotsns
sns.barplot(data=tips, x='day', y='total_bill', estimator=np.mean)
plt.title('Barplot â€” average bill per day')
plt.show()


sns.countplot(data=tips, x='day')
plt.title('Countplot â€” bills per day')
plt.show()
```

```python
#histogramsns
sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.title('Histogram + KDE â€” total bill')
plt.show()
```

```python
#boxplotsns
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Boxplot â€” total bill by day')
plt.show()


sns.violinplot(data=tips, x='day', y='total_bill')
plt.title('Violinplot â€” total bill by day')
plt.show()
```

```python
#pairplotsns
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()


sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='scatter')
plt.show()
```

```python
#implotsns
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', height=5)
plt.show()


sns.catplot(data=tips, x='day', y='total_bill', kind='box', height=4)
plt.show()
```

```python
#heatmapsns
pivot = tips.pivot_table(index='day', columns='time', values='total_bill', aggfunc='mean')
sns.heatmap(pivot, annot=True, fmt='.2f')
plt.title('Heatmap â€” avg bill per day/time')
plt.show()
```
